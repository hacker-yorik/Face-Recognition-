```
def load_vid_to_np(file_name):
  videodata = skvideo.io.vread("/content/gdrive/My Drive/"+file_name+".mp4")
  np_video=np.asarray(videodata)
  return np_video
  

def rescale_np_to_PIL(np_pic,dims=(0,400,1080,1480)):
  my_face_pillow=PIL.Image.fromarray(np_pic)
  my_face_pillow=my_face_pillow.crop(dims).resize((120,120))
  return my_face_pillow

def save_pic_from_vid(Person,file_name,dims=(0,400,1080,1480)):
  my_face_rescaled_np=np.array(())
  i=0
  name="/content/gdrive/My Drive/"+Person+"/"+Person
  for my_face in load_vid_to_np(file_name):
    my_face_pillow=rescale_np_to_PIL(my_face,dims)
    my_face_pillow.save(name+"%04d"%i+".jpg")
    i+=1
```
